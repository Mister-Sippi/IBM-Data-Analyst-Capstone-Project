# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt

QUERY = """
SELECT ConvertedComp
FROM master
GROUP BY ConvertedComp
ORDER BY ConvertedComp
"""
df = pd.read_sql_query(QUERY,conn)

# Plot a histogram of the ConvertedComp column
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted Compensation (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
import matplotlib.pyplot as plt

QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
df = pd.read_sql_query(QUERY,conn)

# Create a boxplot for the Age column with Age on the y-axis
plt.figure(figsize=(8, 6))
plt.boxplot(df['Age'].dropna(), vert=True)
plt.title('Boxplot of Age')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
import matplotlib.pyplot as plt

QUERY = """
SELECT Age,WorkWeekHrs
FROM master
"""
df = pd.read_sql_query(QUERY,conn)

# Plot a scatter plot of Age and WorkWeekHrs
plt.figure(figsize=(20, 10))
plt.scatter(df['Age'], df['WorkWeekHrs'], color='blue', alpha=0.6)
plt.title('Scatter Plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import matplotlib.pyplot as plt

QUERY = """
SELECT Age,WorkWeekHrs,CodeRevHrs
FROM master
"""
df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(20, 8))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age'], alpha=0.6)
plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs (Age as Bubble Size)')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT *, COUNT(*) AS Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5

"""
df = pd.read_sql_query(QUERY,conn)

print(df)

data = df['Count']
labels = df['DatabaseDesireNextYear']

plt.rcParams['figure.figsize'] = [5, 5]

fig, ax = plt.subplots()
ax.pie(data, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
#In the list of most popular languages respondents wish to learn next year, what is the rank of Python?
QUERY = """
SELECT *, COUNT(*) AS Count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY count DESC
LIMIT 5

"""
df = pd.read_sql_query(QUERY,conn)

print(df)

data = df['Count']
labels = df['LanguageDesireNextYear']

plt.rcParams['figure.figsize'] = [5, 5]

fig, ax = plt.subplots()
ax.pie(data, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
#How many respondents indicated that they currently work with 'SQL'?
QUERY = """
SELECT *, COUNT(*) AS Count
FROM LanguageWorkedWith
GROUP BY LanguageWorkedWith
ORDER BY count DESC
LIMIT 5

"""
df = pd.read_sql_query(QUERY,conn)

print(df)

data = df['Count']
labels = df['LanguageWorkedWith']

plt.rcParams['figure.figsize'] = [5, 5]

fig, ax = plt.subplots()
ax.pie(data, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
#How many respondents indicated that they work on 'MySQL' only?
QUERY = """
SELECT *
FROM DatabaseWorkedWith
ORDER BY Respondent
"""
df = pd.read_sql_query(QUERY,conn)

counts = df.groupby('Respondent')['DatabaseWorkedWith'].nunique()
uniques = counts[counts == 1].index

uniq_mysql = df[(df['Respondent'].isin(uniques)) & (df['DatabaseWorkedWith'] == 'MySQL')].shape[0]
print(uniq_mysql)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35

"""
df = pd.read_sql_query(QUERY,conn)

median_hours = df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()
median_hours.plot(kind='bar', stacked=True, figsize=(10, 8))
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.xticks(rotation=0)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60

"""
df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10, 6))
sns.lineplot(data=df.groupby('Age')['ConvertedComp'].median().reset_index(), x='Age', y='ConvertedComp', marker='o')
plt.title('Median Converted Compensation for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.grid(True)
plt.show()
```
